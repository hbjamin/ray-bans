```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
import os

# Get the list of CSV files in the directory
csv_files = glob.glob('*.csv') 
dfs = []  # List to store dataframes and filenames

# Read all CSV files, reverse the DataFrame, and store them in the list with their filenames
for file in csv_files:
    df = pd.read_csv(file)
    
    # Ensure x and y values are numeric
    df.iloc[:, 0] = pd.to_numeric(df.iloc[:, 0], errors='coerce')  # Ensure x values are numeric (Wavelength)
    df.iloc[:, 1] = pd.to_numeric(df.iloc[:, 1], errors='coerce')  # Ensure y values are numeric (T%)
    
    dfs.append((df.iloc[::-1], os.path.basename(file)))  # Store the reversed DataFrame with its filename

# Create a new plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
ax[0].plot(dfs[2][0].iloc[:, 0], dfs[2][0].iloc[:, 1], color='red', linestyle='-', label="$\\theta=0$\u00B0 (Vertical Polarizer)")
ax[0].plot(dfs[3][0].iloc[:, 0], dfs[3][0].iloc[:, 1], color='black', linestyle='-', label="$\\theta=45$\u00B0")
ax[0].plot(dfs[1][0].iloc[:, 0], dfs[1][0].iloc[:, 1], color='blue', linestyle='-', label="$\\theta=90$\u00B0 (Horizontal Polarizer)")
ax[1].plot(dfs[5][0].iloc[:, 0], dfs[5][0].iloc[:, 1], color='red', linestyle='-', label="$\\theta=0$\u00B0 (Vertical Polarizer)")
ax[1].plot(dfs[0][0].iloc[:, 0], dfs[0][0].iloc[:, 1], color='black', linestyle='-', label="$\\theta=45$\u00B0")
ax[1].plot(dfs[4][0].iloc[:, 0], dfs[4][0].iloc[:, 1], color='blue', linestyle='-', label="$\\theta=90$\u00B0 (Horizontal Polarizer)")
# Add titles and labels
ax[0].set_title("Mark's Vertically Polarized + UV Protection Ray-Ban")
ax[1].set_title("Fred's Vertically Polarized + UV Protection Ray-Ban")
ax[0].set_xlabel('Wavelength [nm]')
ax[0].set_ylabel('T%')
# Set the custom limits for x and y axis based on actual values
ax[0].set_xlim(200, 800)  # X-axis from 200 to 800 (actual wavelength values)
ax[0].set_ylim(0, 100)    # Y-axis from 0 to 100 (actual T% values)
ax[1].set_xlim(200, 800)  # X-axis from 200 to 800 (actual wavelength values)
ax[1].set_ylim(0, 100)    # Y-axis from 0 to 100 (actual T% values)
# Add a legend with the filenames
ax[0].legend()
ax[1].legend()
# Show the plot
plt.show();

# Create a new plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
ax[0].plot(dfs[2][0].iloc[:, 0], dfs[2][0].iloc[:, 1], color='red', linestyle='-', label="$\\theta=0$\u00B0 (Vertical Polarizer)")
ax[0].plot(dfs[3][0].iloc[:, 0], dfs[3][0].iloc[:, 1], color='black', linestyle='-', label="$\\theta=45$\u00B0")
ax[0].plot(dfs[1][0].iloc[:, 0], dfs[1][0].iloc[:, 1], color='blue', linestyle='-', label="$\\theta=90$\u00B0 (Horizontal Polarizer)")
ax[1].plot(dfs[5][0].iloc[:, 0], dfs[5][0].iloc[:, 1], color='red', linestyle='-', label="$\\theta=0$\u00B0 (Vertical Polarizer)")
ax[1].plot(dfs[0][0].iloc[:, 0], dfs[0][0].iloc[:, 1], color='black', linestyle='-', label="$\\theta=45$\u00B0")
ax[1].plot(dfs[4][0].iloc[:, 0], dfs[4][0].iloc[:, 1], color='blue', linestyle='-', label="$\\theta=90$\u00B0 (Horizontal Polarizer)")
# Add titles and labels
ax[0].set_title("Mark's Vertically Polarized + UV Protection Ray-Ban")
ax[1].set_title("Fred's Vertically Polarized + UV Protection Ray-Ban")
ax[0].set_xlabel('Wavelength [nm]')
ax[0].set_ylabel('T%')
# Set the custom limits for x and y axis based on actual values
ax[0].set_yscale('log')
ax[1].set_yscale('log')
ax[0].set_xlim(200, 810)  # X-axis from 200 to 800 (actual wavelength values)
ax[0].set_ylim(0, 100)    # Y-axis from 0 to 100 (actual T% values)
ax[1].set_xlim(200, 810)  # X-axis from 200 to 800 (actual wavelength values)
ax[1].set_ylim(0, 100)    # Y-axis from 0 to 100 (actual T% values)
# Add a legend with the filenames
ax[0].legend()
ax[1].legend()
# Show the plot
plt.show();


```
